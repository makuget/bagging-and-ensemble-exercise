```python
from sklearn.preprocessing import StandardScaler

# Separate features (X) and target (y)
X = df_encoded.drop('HeartDisease', axis=1)
y = df_encoded['HeartDisease']

# Apply StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Standalone SVM
svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)

print("Standalone SVM Accuracy:", accuracy_score(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

```python

# SVM with Bagging
bagging_svm = BaggingClassifier(estimator=SVC(), n_estimators=10, random_state=42)
bagging_svm.fit(X_train, y_train)
y_pred_bag_svm = bagging_svm.predict(X_test)

print("Bagged SVM Accuracy:", accuracy_score(y_test, y_pred_bag_svm))
print(classification_report(y_test, y_pred_bag_svm))
```

```python
from sklearn.tree import DecisionTreeClassifier

# Standalone Decision Tree
tree_model = DecisionTreeClassifier(random_state=42)
tree_model.fit(X_train, y_train)
y_pred_tree = tree_model.predict(X_test)

print("Standalone Decision Tree Accuracy:", accuracy_score(y_test, y_pred_tree))
print(classification_report(y_test, y_pred_tree))
```

```python
# Decision Tree with Bagging
bagging_tree = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=10, random_state=42)
bagging_tree.fit(X_train, y_train)
y_pred_bag_tree = bagging_tree.predict(X_test)

print("Bagged Decision Tree Accuracy:", accuracy_score(y_test, y_pred_bag_tree))
print(classification_report(y_test, y_pred_bag_tree))
```

```python
### 💡 Analysis: When to Use Bagging?

- **SVMs** are already stable models — Bagging may not improve much.
- **Decision Trees** are unstable — Bagging can significantly help reduce variance and improve performance.

📌 **Bagging is best when:**
- Base models are high-variance (e.g., decision trees).
- You want to reduce overfitting.
- You have enough data to allow training multiple models.
```
